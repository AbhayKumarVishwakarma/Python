### Assignment 15: Generator Pipeline

Write three generator functions: `integers` that yields integers from 1 to 10, `doubles` that yields each integer doubled, and `negatives` that yields the negative of each doubled value. Chain these generators to create a pipeline that produces the negative doubled values of integers from 1 to 10.

```python
#1

class Countdown:
    def __init__(self, start):
        self.start = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.start <= 0:
            raise StopIteration
        else:
            self.start -= 1
            return self.start
        
for i in Countdown(5):
    print(i)
```

```python
#2

class MyRange:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.start >= self.end:
            raise StopIteration
        else:
            self.start += 1
            return self.start - 1
        
for i in MyRange(1, 10):
    print(i)
```

```python
#3

def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci(10):
    print(num)
```

```python
#4

square = (x**2 for x in range(1, 11))

for num in square:
    print(num)
```

```python
#5

def even_numbers(limit):
    for i in range(limit + 1):
        if i % 2 == 0:
            yield i

def squares(limit):
    for i in limit:
        yield i**2

even_gen = even_numbers(10)
square_gen = squares(even_gen)

for num in square_gen:
    print(num)

```

```python
#6

import time

def time_it(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Execution time: {end_time - start_time} seconds!")
        return result
    return wrapper

@time_it
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
    
print(factorial(25))
```

```python
#7

def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(5)
def welcome():
    print("Welcome to Home")

welcome()
```

```python
#8

def uppercase(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result.upper()
    return wrapper

def exclaim(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result + "!"
    return wrapper

@uppercase
@exclaim
def greet(name):
    return f"Hello, {name}"

greet("Abhay")
```

```python
#9

def singleton(cls):
    instances = {}
    def get_instance(*args, **kwargs):
        if cls not in instances:
            instances[cls] = cls(*args, **kwargs)
        return instances[cls]
    return get_instance

@singleton
class DatabaseConnection:
    def __init__(self):
        print("Database connection created!")

db1 = DatabaseConnection()
db2 = DatabaseConnection()

print(db1 is db2)
```
