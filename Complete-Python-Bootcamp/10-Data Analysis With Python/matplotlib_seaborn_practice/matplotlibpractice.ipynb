```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 8, 10, 8, 6]

plt.plot(x, y, '.-')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')
plt.show()
```

```python
# Plotting

x = np.array([1, 5, 3, 7, 9])
y = np.array([2, 8, 10, 8, 6])

plt.plot(x, y, marker ='o', color='b', ls='--', lw=3, ms=20, mec='g', mfc='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')
plt.grid(True)
plt.show()
```

```python
# Labels and Title

x = np.array([1, 5, 3, 7, 9])
y = np.array([2, 8, 10, 8, 6])

font = {'family':'serif','color':'darkred','size':15}

plt.plot(x, y, '.-')
plt.xlabel('X axis', fontdict=font)
plt.ylabel('Y axis', fontdict=font)
plt.title('Line Plot', loc='left', fontsize=20, fontweight='bold', color='g')
plt.grid(True)
plt.show()
```

```python
# Grid

x = np.array([1, 5, 3, 7, 9])
y = np.array([2, 8, 10, 8, 6])

plt.plot(x, y, '.-')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Plot')
plt.grid(axis='x', color='b', ls='--', lw=2)
plt.show()
```

```python
# Subplot

x = np.array([1, 5, 3, 7, 9])
y = np.array([2, 8, 10, 8, 6])

plt.subplot(2, 2, 1)
plt.plot(x, y, '.-')

x = np.array([1, 5, 3, 7, 9])
y = np.array([2, 8, 10, 8, 6])

plt.subplot(2, 2, 2)
plt.plot(x, y, '.-')

x = np.array([1, 5, 3, 7, 9])
y = np.array([2, 8, 10, 8, 6])

plt.subplot(2, 2, 3)
plt.plot(x, y, '.-')

x = np.array([1, 5, 3, 7, 9])
y = np.array([2, 8, 10, 8, 6])

plt.subplot(2, 2, 4)
plt.plot(x, y, '.-')

plt.show()
```

```python
x1 = np.array([1, 5, 3, 7, 9])
y1 = np.array([2, 8, 10, 8, 6])

x2 = np.array([1, 5, 3, 8, 9])
y2 = np.array([2, 3, 10, 8, 6])

plt.subplot(2, 1, 1)
plt.plot(x1, y1, '.-', color='b')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-', color='r')

plt.show()
```

```python
# Scatter Plot

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')

plt.colorbar()
plt.show()
```

```python
x = np.random.randint(100, size=100)
y = np.random.randint(100, size=100)

colors = np.random.randint(100, size=100)
sizes = np.random.randint(100, size=100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

plt.colorbar()

plt.show()
```

```python
# Bar Plot

x = np.array(['A', 'B', 'C', 'D', 'E'])
y = np.array([3, 8, 1, 10, 5])

plt.bar(x, y, width=0.5, edgecolor='b', color='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar Plot')
plt.show()
```

```python
# Bar horizontal

x = np.array(['A', 'B', 'C', 'D', 'E'])
y = np.array([3, 8, 1, 10, 5])

plt.barh(x, y, height=0.1, edgecolor='b', color='r')
plt.show()
```

```python
# Histograms

x = np.random.normal(170, 10, 250)
plt.hist(x, 100)
plt.title('Histogram')
plt.show()
```
